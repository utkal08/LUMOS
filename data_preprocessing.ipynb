```python
import os
import spacy

def process_text_files(directory_path):
    processed_data = []  # Initialize an empty list to store processed data

    
    files = [file for file in os.listdir(directory_path) if file.endswith('.txt')]

    
    nlp = spacy.load("en_core_web_sm")

    for file in files:
        file_path = os.path.join(directory_path, file)

        # Read the data from the current file
        with open(file_path, "r", encoding="utf-8") as file_content:
            data = file_content.read()

        # Clean and preprocess the data using SpaCy
        doc = nlp(data)

        # Store the processed data (or tokens, embeddings, etc.) in the list
        processed_data.append(doc)  

    return processed_data

# Specify the directory containing your .txt files
directory_path = r"C:\Users\utkal\LUMOS"

# Call the function to process the text files in the specified directory
processed_data_list = process_text_files(directory_path)

# Now, processed_data_list contains processed data for all the text files in the specified directory
```
